```python
from pymongo import MongoClient
from dotenv import load_dotenv
import os

load_dotenv()

SSH_HOST = os.getenv("SSH_HOST")
SSH_PORT = int(os.getenv("SSH_PORT", 22))
SSH_USER = os.getenv("SSH_USER")
SSH_KEY = os.getenv("SSH_KEY")

# MongoDB 설정
MONGO_HOST = os.getenv("MONGO_HOST", "localhost")
MONGO_PORT = int(os.getenv("MONGO_PORT", 27017))
MONGO_USER = os.getenv("MONGO_USER")
MONGO_PASSWORD = os.getenv("MONGO_PASSWORD")
MONGO_AUTH_DB = os.getenv("MONGO_AUTH_DB", "admin")

import warnings
warnings.filterwarnings("ignore", module="paramiko")
from sshtunnel import SSHTunnelForwarder


server = SSHTunnelForwarder(
    (SSH_HOST, SSH_PORT),
    ssh_username=SSH_USER,
    ssh_pkey=SSH_KEY,
    remote_bind_address=(MONGO_HOST, MONGO_PORT)
)
server.start()

client = MongoClient(
    f"mongodb://{MONGO_USER}:{MONGO_PASSWORD}"
    f"@127.0.0.1:{server.local_bind_port}/?authSource={MONGO_AUTH_DB}"
    )

print(client.list_database_names())
```

```python
db = client['class']
student = db['student']
```

```python
# 필드를 특정 값으로 설정하기
student.update_one(
    {"name": "Alice"},
    {"$set": {"score": 96, "major": 'EE'}}
)
```

```python
# 필드 더하기 
student.update_one(
    {"name": "Alice"},
    {"$inc": {"age": 1}}
)
```

```python
# 필드 감소하기
student.update_one(
    {"name": "Alice"},
    {"$inc": {"score": -2}}
)
```

```python
# 필드에 곱하기
student.update_one(
    {"name": "Alice"},
    {"$mul": {"score": 1.1}}
)
```

```python
# 리스트 만들기
student.update_one(
    {"name": "Alice"},
    {"$set": {
        "devices": []
    }}
)
```

```python
# 리스트에 값 1개 추가하기
student.update_one(
    {"name": "Alice"},
    {"$push": {"devices": "iPad"}}
)
```

```python
# 리스트에 값 여러 개 추가하기
student.update_one(
    {"name": "Alice"},
    {
        "$push": {
            "devices": {
                "$each": ["MacBook", "iPhone"]
            }
        }
    }
)
```

```python
# 리스트에 중복 없이 1개 추가하기
student.update_one(
    {"name": "Alice"},
    {
        "$addToSet": {
            "devices": "iPhone"
        }
    }   
)
```

```python
# 리스트에 중복 없이 여러 개 추가하기
student.update_one(
    {"name": "Alice"},
    {
        "$addToSet": {
            "devices": {
                "$each": ["iPad", 'iPhone', 'iMac']
            }
        }
    }   
)
```

```python
# 리스트 필드 추가
student.update_one(
    {"name": "Alice"},
    {
        "$set": {
            "scores": [50, 60, 70, 80, 90, 100]
        }
    }
)
```

```python
# 리스트에서 제거
student.update_one(
    {"name": "Alice"},
    {"$pull": {
        "scores": 50
    }}
)
```

```python
# 조건을 줘서 제거(pull)
student.update_one(
    {"name": "Alice"},
    {"$pull": {
        "scores": {"$gte": 90}
    }}
)
```
