```python

while True:
    ans = 28
    anticipation = int(input('숫자를 입력하세요(1~50) : '))
    if anticipation > ans:
        print('Down!')
        continue
    elif anticipation < ans:
        print('Up!')
        continue
    elif anticipation == ans:
        print('정답입니다!')
        break
```

```python
scores = [85, 45, 70, 0, 95, 55] 
consequence = 0
while True:
    for score in scores:
        if score == 0:
            consequence = 0
            print('0점 발견! 채점을 중단합니다.')
            break
        elif score >= 60:
            consequence += score
        elif score < 60:
            continue
    break
print(f'최종 합계 : {consequence}')
```

```python
balance = 10000
while True:
    withdrawl = input('출금할 금액을 입력하세요 (종료 : 0000) : ')
    if withdrawl == '0000':
        break
    withdrawl = int(withdrawl)
    if withdrawl > balance:
        print('잔액이 부족혀 이눔시끼')
        continue
    elif withdrawl <= balance:
        balance -= withdrawl
        print(f'{withdrawl}원이 출금되었습니다. 남은 잔액 : {balance}원')
        if balance == 0:
            print('잔액을 모두 소진하였습니다.')
            break
        continue
    
```

```python
total = 0
while True:
    client = int(input('입장 인원siuuuuuuuuu를 입력하세요: '))
    if client == 0:
        print('강제 종료혈게')
        break
    elif (client + total) > 10:
        print('입장 인원 수가 넘 마늠.')
        continue
    else:
        total += client
        print(f'현재 인원 훠훠 : {total}/10')
        if total == 10:
            print('만석.')
            break
```

```python
reservation = []
while True:
    ticket = int(input('예약할 좌석 번호를 입력하세요(1~10 / 종료:0) : '))
    if ticket in reservation:
        print('이미 예약된 좌석입니다.')
        continue
    elif ticket == 0:
        print('예약 시스템을 종료합니다.')
        break
    elif len(reservation) == 10:
        print('모든 좌석이 매진되었습니다.')
        break
    else:
        reservation.append(ticket)
        print(f'{ticket}번 좌석 예약 완료 (현재 {len(reservation)})')
        continue
```

```python
class Student:
    def __init__(self, score, name):
        self.score = score
        self.name = name

student_list = [Student(85, '철수'), Student(92, '영희'), Student(45, '민수'), Student(0, '지siuuuuu')]

for student in student_list:
    if student.score >= 80:
        print(f'{student.name}({student.score}점) : 우수')
    elif student.score == 0:
        print(f'경고: {student.name} 학생의 성적이 누락되었습니다. 점검을 중단합니다')
    elif student.score < 60:
        print(f'{student.name}({student.score}점) : 보충학습 대상 및 ㅄ')
    


```

```python
class VendingMochine:
    def __init__(self):
        self.recon = 5

    def push(self):
        self.recon -= 1

japangi = VendingMochine()
while True:
    option = input('버튼을 누르세요 (push / 종료: quit):')
    if option.lower() == 'push':
        japangi.push()
        print(f'음료가 나왔습니다.(현재 재고:{japangi.recon})')
        if japangi.recon == 2:
            print('!! 알림: 재고가 얼마 남지 않았습니다 !!')
        elif japangi.recon == 0:
            print('품절되었습니다. 영업을 종료합니다.')
            break
    elif option.lower() == 'quit':
        break
```

```python
class Sukill:  
    def __init__(self, name, mp):
        self.name = name
        self.mp = mp

skills = [
    Sukill("평타", 0),
    Sukill("파이어볼", 20),
    Sukill("메테오", 60),  # 마나 부족으로 건너뜀
    Sukill("궁극기", 30),
    Sukill("연속 베기", 10) # 궁극기 이후라 실행 안 됨
]
cur_mp = 50
for skill in skills:
    if skill.name == '궁극기':
        cur_mp == 0
        print('궁극기 시전!(남은 MP:0)')
        print('모든 힘을 쏟아부었습니다! 전투 종료.')
        break
    elif skill.mp <= cur_mp:
        cur_mp -= skill.mp
        print(f'{skill.name} 시전! (남은 MP : {cur_mp - skill.mp})')
    else:
        print(f'마나가 부족하여 {skill.name} 스킬을 사용할 수 없습니다.')
        
```

```python
class Machine:
    def __init__(self, name):
        self.name = name
        self.temp = 20
    def work(self):
        self.temp += 10
ROBOT = Machine('A-1 로봇')
for i in range(0, 4):
    ROBOT.work()
    print(f'{i}번째 공정 시작')
    print(f'{ROBOT.name} 가동 중... 현재 온도: {ROBOT.temp}')
    if ROBOT.temp >= 60:
        print('경고: 기계 과열! 작업을 강제 종료합니다.')
```

```python
class Food:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def get_price(self):
        return self.price
cart = [
    Food("떡볶이", 5000),
    Food("치킨", 18000),
    Food("스테이크", 35000),
    Food("콜라", 2000)
]
price = 0
for food in cart:
    if food.get_price() >= 20000:
        print(f'[{food.name}]은 너무 비싸서 제외합니다.')
        continue
    price += food.get_price()
    print(f'{food.name} 추가 완료 (현재 합계: {price})')
print(f'최종 결제 금액 : {price}')
    
```

```python
class Character:
    def __init__(self):
        self.level = 1
        self.experience = 0
    
    def hunt(self):
        self.experience += 40
        if self.experience >= 100:
            self.experience == 0
            self.level += 1
            print(f'★ 축하합니다! 레벨업! (현재 레벨: {self.level}) ★')
            if self.level == 3:
                print('목표 레벨 달성! 전직이 가능합니다.')
    
Jihoo = Character()
while True:
        hunt = input('사냥하시겠습니까? (Y/n) : ')
        if hunt.lower() == 'y':
            Jihoo.hunt()
            print(f'사냥 성공! 경험치 : +40(현재 경험치 : {Jihoo.experience})')
        else:
            break

```

```python
class Car:
    def __init__(self, VIN, is_electric):
        self.VIN = VIN
        self.is_electric = is_electric

parking_tower = [
    Car("12가1234", True),
    None,
    Car("34나5678", False), # 전기차 구역에 일반차 주차됨
    Car("56다9012", True)
]
for num, i in enumerate(parking_tower):
    if not i:
        print(f"{num}번 구역 : 비어 있음.")
    elif i.is_electric == False:
        print(f'비상! {num}번 전기차 구역에 일반 차량 {i.VIN} 발견! 단속팀 출동.')
        break
    else:
        print(f"{num}번 구역 : 차량 {i.VIN}")
```
