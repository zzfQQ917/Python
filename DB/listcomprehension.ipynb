```python
nums = [1, 2, 3]
result = []
for n in nums:
    if n % 2 == 0:
        result.append('even')
    else:
        result.append('odd')
result = ['even' if n % 2 == 0 else 'odd' for n in nums]
```

```python

while True:
    ans = 28
    anticipation = int(input('숫자를 입력하세요(1~50) : '))
    if anticipation > ans:
        print('Down!')
        continue
    elif anticipation < ans:
        print('Up!')
        continue
    elif anticipation == ans:
        print('정답입니다!')
        break
```

```python
scores = [85, 45, 70, 0, 95, 55] 
consequence = 0
while True:
    for score in scores:
        if score == 0:
            consequence = 0
            print('0점 발견! 채점을 중단합니다.')
            break
        elif score >= 60:
            consequence += score
        elif score < 60:
            continue
    break
print(f'최종 합계 : {consequence}')
```

```python
balance = 10000
while True:
    withdrawl = input('출금할 금액을 입력하세요 (종료 : 0000) : ')
    if withdrawl == '0000':
        break
    withdrawl = int(withdrawl)
    if withdrawl > balance:
        print('잔액이 부족혀 이눔시끼')
        continue
    elif withdrawl <= balance:
        balance -= withdrawl
        print(f'{withdrawl}원이 출금되었습니다. 남은 잔액 : {balance}원')
        if balance == 0:
            print('잔액을 모두 소진하였습니다.')
            break
        continue
    
```

```python
items = ['normal', 'sale', 'normal', 'normal', 'broken', 'normal']
price = 0
for item in items:
    if item == 'normal':
        price += 1000
    elif item == 'sale':
        continue
    else:
        print('파손 물품 발견1 검tn을 중단합니다.')
        print(f'최종 계산 금액: {price}원')
        break
```

```python
total = 0
while True:
    client = int(input('입장 인원siuuuuuuuuu를 입력하세요: '))
    if client == 0:
        print('강제 종료혈게')
        break
    elif (client + total) > 10:
        print('입장 인원 수가 넘 마늠.')
        continue
    else:
        total += client
        print(f'현재 인원 훠훠 : {total}/10')
        if total == 10:
            print('만석.')
            break
```

```python
for num in range(2, 10):
    if (num % 2) != 0:
        continue
    print(f'---{num}단---')
    for times in range(1, 10):
        print(f'{num} x {times} = {num*times}')
        if times >= num:
            break
```

```python
list_a = [1, 2, 3]
list_b = [2, 3, 4]
for a in list_a:
    for b in list_b:
        if a == b:
            print(f'짝 발견: {a} == {b}')
            break
        else:
            print(f'{a}의 짝은 없습니다.')
            break
```

```python
reservation = []
while True:
    ticket = int(input('예약할 좌석 번호를 입력하세요(1~10 / 종료:0) : '))
    if ticket in reservation:
        print('이미 예약된 좌석입니다.')
        continue
    elif ticket == 0:
        print('예약 시스템을 종료합니다.')
        break
    elif len(reservation) == 10:
        print('모든 좌석이 매진되었습니다.')
        break
    else:
        reservation.append(ticket)
        print(f'{ticket}번 좌석 예약 완료 (현재 {len(reservation)})')
        continue
```
