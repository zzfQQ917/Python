```python
var1 = int(input('입력!: '))
square = 1
allStairs = []

if var1//2 != 0:
    space = var1 - 1
    for i in range((var1+1)//2):
        newList = []
        for j in range(square):
            newList.append('■')
        
        for k in range(space):
            newList.append(' ')
        
        for l in range(square):
            newList.append('■')
        allStairs.append(newList)
        square += 1
        space -= 2

else:
    space = var1
    for i in  range(var1//2):
        newList = []
        for j in range(square):
            newList.append('■')
        
        for k in range(space):
            newList.append(' ')
        
        for l in range(square):
            newList.append('■')
        allStairs.append(newList)
        square += 1
        space -= 2

for stair in allStairs:
    print(stair)
```
