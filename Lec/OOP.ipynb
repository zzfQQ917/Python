```python
# 1
class Menu:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        self.orders = []
    
    def order(self, customerName):
        if customerName in self.orders:
            print(f"{customerName} → 이미 주문하셨습니다. ")
        else:
            print(f"{customerName} → 주문 완료! ")
            self.orders.append(customerName)
    
    def cancelOrder(self, customerName):
        if customerName not in self.orders:
            print(f"{customerName} → 주문 기록이 없습니다. ")
        else:
            print(f"{customerName} → 주문 취소 완료! ")
            self.orders.remove(customerName)
    
    def showOrders(self):
        print(self.orders)
        print(f"{self.name} 주문자: {self.orders}")
        
    
iceAmericano = Menu("아메리카노", 4000)
caffeLatte = Menu("카페라떼", 4500)

iceAmericano.order("홍길동")
caffeLatte.order("김영희")

iceAmericano.cancelOrder("홍길동")
caffeLatte.cancelOrder("김영희")

iceAmericano.showOrders()
caffeLatte.showOrders()





        
```

```python
class Movie:
    totalSeats = 10
    totalReserved = 0
    def __init__(self, title):
        self.title = title
        self.reservations = []
    
    def reserve(self, name):
        if Movie.totalSeats == Movie.totalReserved:
            print("→ 예약이 불가합니다. ")
        elif name in self.reservations:
            print("→ 이미 예약하셨습니다. ")
        else:
            print("→ 예약 완료!")
            Movie.totalReserved += 1
            self.reservations.append(name)
    
    def cancel(self, name):
        if name not in self.reservations:
            print("→ 예약 기록이 업써용. ")
        else:
            print("→ 예약 취소 완료! ")
            self.reservations.remove(name)
            Movie.totalReserved -= 1
    
    def showReservations(self):
        print(f"{self.title} 예약자: {self.reservations}")
    
avengersSaga = Movie("어벤져스 ")
Inception = Movie("인셉션 ")

avengersSaga.reserve("홍길동 ")
Inception.reserve("김영희 ")
Inception.reserve("홍길동 ")

Inception.cancel("김영희 ")

avengersSaga.showReservations()
Inception.showReservations()
print(f"총 예약 수: {Movie.totalReserved} / 10")


 


```

```python
class Book:
    totalReserved = 0

    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.reservedBy = []

    def reserve(self, name):
        if name in self.reservedBy:
            print("→ 이미 예약하셨슈. ")
        else:
            print("→ 예약 완료여! ")
            self.reservedBy.append(name)
            Book.totalReserved += 1
    def cancel(self, name):
        if name not in self.reservedBy:
            print("→ 예약 기록이 없습니다. ")
        else:
            print("→ 예약 취소 완료여! ")
            self.reservedBy.remove(name)
            Book.totalReserved -= 1

    def showReservations(self):
        print(f"{self.title} 예약자: {self.reservedBy}")

firstBook = Book("운영체제", "홍길동")
secondBook = Book("자료구조", "김영희")

firstBook.reserve("이순신 ")
secondBook.reserve("강감찬 ")
secondBook.reserve("이순신 ")
secondBook.cancel("강감찬 ")

firstBook.showReservations()
secondBook.showReservations()

print(f"전체 예약 수: {Book.totalReserved}")
        
```

```python
class Employee:
    def __init__(self, name):
        self.name = name
    
    def showInfo(self):
        print(f"이름: {self.name}")
    
class FullTimeEmployee(Employee):
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary

    def getMonthlyPay(self):
        print(f"월급({self.salary / 12}) ")

class PartTimeEmployee(Employee):
    def __init__(self, name, hourlyWage, hours):
        super().__init__(name)
        self.hourlyWage = hourlyWage
        self.hours = hours
    
    def getMonthlyPay(self):
        print(f"월月급({self.hourlyWage * self.hours})")

hired1 = FullTimeEmployee("홍길동吉洞", 48000000)
hired2 = PartTimeEmployee("김영희", 1500, 80)

hired1.showInfo()
hired2.showInfo()

hired1.getMonthlyPay()
hired2.getMonthlyPay()
```

```python
class Device:
    def __init__(self, name):
        self.name = name
        self.battery = 0
    
    def show_status(self):
        print(f"기기: {self.name} / 배터리: {self.battery} %")
    
class Phone(Device):
    def charge(self):
        self.battery += 20
    
class Laptop(Device):
    def charge(self):
        self.battery += 30

cell_Phone = Phone("스마트폰")
device_laptop = Laptop("노트북")

cell_Phone.charge()
cell_Phone.charge()

device_laptop.charge()
device_laptop.charge()

cell_Phone.show_status()
device_laptop.show_status()

```

```python
class Mob:
    total_attacks = 0
    def __init__(self, name):
        self.name = name
        self.hp = 100
    
    def show_status(self):
        print(f"몬스터: {self.name} / 체력: {self.hp}")

    def show_total_attacks():
        print(f"전체 공격 횟수: {Mob.total_attacks}회")
    
class Zombie(Mob):
    def __init__(self, name, level):
        super().__init__(name)
        self.level = level
    
    def attack(self):
        print(f"{self.name} (레벨 {self.level}) (이)가 공격합니다 ! (대미지: 10)")
        self.hp -= 5
        Mob.total_attacks += 1

class Creeper(Mob):
    def __init__(self, name, power):
        super().__init__(name)
        self.power = power

    def attack(self):
        print(f"{self.name} (폭발력 {self.power}) (이)가 폭발합니다 ! (대미지: 30)")
        self.hp -= 20
        Mob.total_attacks += 1

spawned_zombie = Zombie("좀비A", 3)
spawned_creeper = Creeper("좀비B", 5)

spawned_zombie.attack()
spawned_creeper.attack()

spawned_zombie.attack()
spawned_creeper.attack()

spawned_zombie.show_status()
spawned_creeper.show_status()

Mob.show_total_attacks()
```
